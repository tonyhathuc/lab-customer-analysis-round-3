## Create a scatterplot between total claim amount and income.
Play around with the parameters of the scatterplot <br>
(markersize? alpha?) and try to identify more features within the data just visually. <br>
You can also try different seaborn plots.

```python
plt.figure(figsize=(16, 10))
#fig,axe=plt.subplots()
plt.title("Total Claim amount vs Income")
claimvsincome = sns.scatterplot(data=data, x="income", y="total_claim_amount",s=200, alpha = .1)
claimvsincome;
```


## Interesting insights from this visualization:
- Strong tendency of claims when the customer has no income
- Empty data under 10000 dollars of wage, the company could have a policy to not accept customers under that threshold
- Horizontal line circa 350 dollar claims, could indicate a "Standard" compensation amount the company is giving out
